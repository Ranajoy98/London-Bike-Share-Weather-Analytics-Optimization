# Dta preprocessing

```python
# Renaming the columns

renamed_dict ={
    'timestamp':'time',
    'cnt':'count_of_a_new_bike_shares', 
    't1':'real_temperature_in_C',
    't2':'temperature_in_C_feels_like',
    'hum':'humidity_in_percentage',
    'wind_speed':'wind_speed_in_kmh',
    'weather_code':'category_of_the_weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}


df.rename(renamed_dict, axis=1, inplace=True)
```

```python
df
```

```python
# changing the humidity values to interger values (0-1) 
df.humidity_in_percentage = df.humidity_in_percentage / 100
```

```python
df
```

```python
# season dictionary to change integers 0-3 to the actual seasons
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

# weather dictionary to map the integers to actual wether catagories
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

# changing the seasons column data type to string
df.season = df.season.astype('str')
# mapping the values 0-3 to seasons
df.season = df.season.map(season_dict)

# changing the weather column data type to string
df.category_of_the_weather = df.category_of_the_weather.astype('str')
# mapping the values to the actual wether catagories
df.category_of_the_weather = df.category_of_the_weather.map(weather_dict)
```
